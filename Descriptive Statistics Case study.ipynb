```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#read the "employee dataset" file as a pandas dataframe
data = pd.read_csv(r"C:\Users\91908\Downloads\Employee Dataset.csv")
#observing the dataset
print('Data:\n',data.head())
#getting the dimensions
print('Dimension:',data.shape)
#getting row names of the dataset
print('Rows:',data.index)
#getting the column names
print('Columns:',data.columns)
#getting basic info
print('Basic Info')
data.info()
```

```python
#visualise the distribution of ratings and compensations(sactter plot, count plot and histogram)
#1.Create a scatter plot of the'age' and 'employee salary' features
plt.figure(figsize = (10,5))
sns.scatterplot(data = data,x = 'age',y = 'salary',color = 'blue',edgecolor = 'linen',alpha = 0.5)
plt.title('Scatterplot of Employee Age v/s Employee Salary')
plt.xlabel('Employee Age')
plt.ylabel('Employee Salary')
plt.show()
#conclusion: As age increases, the salary also increases, though there are some outliers present as well 
```

```python
#2.Create a scatter plot of the'healthy_eating' and 'active_lifestyle' features
plt.figure(figsize = (10,5))
sns.scatterplot(data = data,x = 'healthy_eating',y = 'active_lifestyle',color = 'lime',edgecolor = 'darkgreen',alpha = 0.5)
plt.title('Scatterplot of Healthy eating v/s Active lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
#conclusion: As healthy eating increases, the active lifestyle also increases. 
```

```python
#3.Create a scatter plot of the'healthy_eating' and 'salary' features
plt.figure(figsize = (10,5))
sns.scatterplot(data = data,x = 'healthy_eating',y = 'salary',color = 'purple',edgecolor = 'k',alpha = 0.5)
plt.title('Scatterplot of Healthy eating v/s Salary')
plt.xlabel('Healthy Eating')
plt.ylabel('Salary')
plt.show()
#conclusion: As healthy eating increases, the salary also increases. 
```

```python
#4.Create a count plot of the "groups"feature.
plt.figure(figsize = (10,5))
all_colors = ['#FF6F61', '#6A67CE', '#00C49A', '#FFBB28', '#F75C03']
# Trim colors based on actual unique group count
group_count = data['groups'].nunique()
custom_colors = all_colors[:group_count]
#plot the graph
plt.figure(figsize=(10, 5))
sns.countplot(data=data,x='groups',hue='groups',palette=custom_colors,edgecolor='linen',alpha=0.7,legend=True)
plt.title("Count plot of Groups", fontsize=14, fontweight='bold')
plt.xlabel('Blood Groups')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=1.0)
plt.tight_layout()
plt.show()
# observation : Most of the employees belong to group A and O, With group A having maximum frequency
```

```python
#5.Create a histogram of 'salary' features
plt.figure(figsize = (10,5))
sns.histplot(data = data,x = 'salary',color = 'pink',edgecolor = 'k',alpha = 0.7, bins = 10)
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.grid(axis = 'y',linestyle ='--',alpha = 1.0)
plt.show()
#based on the histogram we can see that the count of highest earning employee is very less 
```
